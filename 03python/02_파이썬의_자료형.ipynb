# 파이썬의 자료형(data type)
## 파이썬의 자료형은 동적으로 변화
* int: 정수(소숫점이 없는 숫자)
* float: 실수(소숫점이 있는 숫자)
* str: 문자/문자열
* list: 여러개의 자료가 순서대로 저장된 자료형, 모든 자료형을 넣어 만들 수 있음
* tuple: list와 비슷하나 한 번 만들면 수정이 안됨, 모든 자료형을 넣을 수 있음
* dictionary(dict): key : value 쌍으로 이루어진 자료형, 모든 자료형을 넣을 수 있음

# 리스트  list
* 리스트는 [정수, 실수, 문자, 리스트, 튜플 , 딕셔너리 ] 모두 넣을 수있다.
* 리스트는 여러 자료가 같이 나열된 형태이므로 순서가 있다.
* 인덱싱, 슬라이싱 가능
* 자료의 추가, 삭제가 가능하다.

```python
l1 = [1, 2.0, '삼', [1,2,3], (4,5,6), {"key":"value"}]
l1
```

```python
print(l1, type(l1))
```

```python
l1[5]
```

```python
l1[:4]
```

```python
l1[2:5]
```

list는 순서가 있는 자료형

```python
l1.append("리스트 자료형")
```

```python
l1
```

```python
l1[6]
```

```python
del l1[6]
```

```python
l1
```

# 리스트는 리스트 내의 자료형을 그대로 유지

```python
for item in l1:
    print(item, type(item))
    
```

* 튜플이 리스트 보다 코드 실행시 더 빠르다.

```python
l3.append("튜플")
```

```python
l3[-1]
```

```python
del l3[-1]
```

```python
l3[2:4]
```

```python
l3[3][1]
```

# 딕셔너리형(dict) JSON {"key" : "value"}
* key와 value로 이루어진 데이터형
* json 이라고 불리며 웹상에서 데이터를 주고 받을 때 주로 사용
* 딕셔너리의 value에는 모든 자료형이 올 수 있음
* 딕셔너리는 순서가 없는 자료형
* key를 호출하면 value가 출력됨

```python
dic1 = {"이름" : "홍길동", "나이" : 33, "키" : 175.4, 
        "가족" : ["둘리", "또치", "도우너"]}
```

```python
dic1["이름"]
```

```python
dic1["나이"]
```

```python
dic1["키"]
```

```python
dic1["가족"]
```

* 도우너만 출력

```python
dic1["가족"][2]
```

```python
dic2 = {"이름" : "홍길동", "나이" : 33, "키" : 175.4, 
        "가족" : [{"둘리": ["초능력", "공룡"]}, {"또치": ["달리기", "타조"]},
                  {"도우너" : ['음악', '외계인']}]}
```

```python
dic2['가족'][2]["도우너"][0]
```

# 형변환 함수
* 형변환: data type 다른 data type으로 바꾸는 것

* 정수로 형변환 int()

* 실수로 변환 float()

* 문자열로 변환 str()

* 리스트 형변환 list()

* 튜플로 형변환 tuple()

# input() : 키보드 입력을 받는 함수
* input("화면에 표시될 메시지")
* 입력 받는 값은 " "으로 구분해서 여러개를 받을 수 있음
* 입력 받는 값은 모두 str 타입의 문자열

* input().split() 으로 여러 데이터를 list 형태로 받기
