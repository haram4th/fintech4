# scikit-learnÏùò one-hot encoderÎ°ú Î≥ÄÌôò ÌõÑ Ïû¨Î∂ÑÏÑù

````python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
````

````python
X_train['age'] = X_train['age'].astype('str')
X_test['age'] = X_test['age'].astype('str')
````

````python
def to_ohe(X_train, X_test):
    cat_cols = X_train.select_dtypes(include="object").columns
    ct = ColumnTransformer(transformers=[('ohe', OneHotEncoder(drop='first', handle_unknown='ignore',
                                                              sparse_output=False), cat_cols)],
                          remainder='passthrough')
    ct.set_output(transform='pandas')
    X_train_final = ct.fit_transform(X_train)
    X_test_final = ct.transform(X_test)
    display(X_train_final.head(2))
    display(X_test_final.head(2))
    print("X_train_columns: ", len(X_train_final.columns))
    print("X_test_columns: ", len(X_test_final.columns))
    return X_train_final, X_test_final
````

````python
X_train_ohe, X_test_ohe = to_ohe(X_train, X_test)
````

# Îç∞Ïù¥ÌÑ∞ Î∂ÑÏÑù Î™®Îç∏Ïùò ÏÑ±Îä•ÌèâÍ∞Ä
1. accuracy(Ï†ïÌôïÎèÑ/Ï†ïÎ∂ÑÎ•òÏú®): 1ÏùÑ 1Ïù¥ÎùºÍ≥† ÎßûÏ∂îÍ≥† 0ÏùÑ 0Ïù¥ÎùºÍ≥† ÎßûÏ∂ò ÎπÑÏú®
2. Error Rate(Ïò§Î∂ÑÎ•òÏú®): 1ÏùÑ 0Ïù¥ÎùºÍ≥† ÏòàÏ∏°ÌïòÍ≥† 0ÏùÑ 1Ïù¥ÎùºÍ≥† ÏòàÏ∏°Ìï¥ÏÑú ÏûòÎ™ª ÏòàÏ∏°Ìïú Í≤É
3. precision(Ï†ïÎ∞ÄÎèÑ): ÏòàÏ∏°Ìïú Í≤É Ï§ëÏóêÏÑú Ïã§Ï†úÎ°ú ÎßûÏùÄ Í≤É, ÏòàÏ∏° postsitve Ï§ëÏóêÏÑú Ïã§Ï†ú TrueÏù∏ ÎπÑÏú®
4. recall(Ïû¨ÌòÑÏú®/ÎØºÍ∞êÎèÑ): Ïã§Ï†úÍ∞íÏùÑ Ïã§Ï†úÍ∞íÏúºÎ°ú Ïûò ÏòàÏ∏°Ìïú Í≤É, Ïã§Ï†ú TrueÏ§ë ÏòàÏ∏° postiveÍ∞Ä ÎßûÏùÄ Í≤É
5. F1-Score: precisionÍ≥º recallÏùò Ï°∞Ìôî ÌèâÍ∑†, 
6. Specificity(ÌäπÏù¥ÎèÑ): Ïã§Ï†ú nagativeÏ§ë ÏòàÏ∏°Ïóê ÏÑ±Í≥µÌïú ÎπÑÏú®

# confusion matrix(ÌòºÎèôÌñâÎ†¨)
|              |  ÏòàÏ∏° Positive    | ÏòàÏ∏° Nagative     |
|:--------------:|:------------------:|:-------------------:|
|**Ïã§Ï†ú True** | TP (True Positive) |FN (False Nagative)|
|**Ïã§Ï†ú False** | FP (False Positive)|TN (True Negative) |
* TP: Ïã§Ï†ú True, ÏòàÏ∏° Positive(Ïã§Ï†ú Ïó∞Î¥â >5Îßå, ÏòàÏ∏° Ïó∞Î¥â >5Îßå)
* TN: Ïã§Ï†ú False, ÏòàÏ∏° Negative(Ïã§Ï†ú Ïó∞Î¥â <=5Îßå, ÏòàÏ∏° Ïó∞Î¥â <=5Îßå)
* FP(1Ìòï Ïò§Î•ò): Ïã§Ï†ú FalseÏù∏Îç∞ PositiveÎ°ú ÏòàÏ∏°(Ïã§Ï†ú Ïó∞Î¥â <=5Îßå, ÏòàÏ∏° Ïó∞Î¥â >5Îßå )
* FN(2Ìòï Ïò§Î•ò): Ïã§Ï†ú TrueÏù∏Îç∞ NegativeÎ°ú ÏòàÏ∏°(Ïã§Ï†ú Ïó∞Î¥â >5Îßå, ÏòàÏ∏° Ïó∞Î¥â <=5Îßå)



* Ï†ïÌôïÎèÑ(accuracy_score): $ TP + TN \over TP+TN+FP+FN $
* Ïò§Î∂ÑÎ•òÏú®(Error Rate): $ FP+FN \over TP+TN+FP+FN $
* Ï†ïÎ∞ÄÎèÑ(precision): $ TP \over TP+FP $
* Ïû¨ÌòÑÏú®/ÎØºÍ∞êÎèÑ(recall): $ TP \over TP+FN $
* ÌäπÏù¥ÎèÑ(Specificity): $ TN \over FP+TN $
* F1-score: $  2 \cdot {Precision \cdot Recall \over Precision + Recall} $

# üìä Î∂ÑÎ•òÎ∂ÑÏÑù ÌèâÍ∞ÄÏßÄÌëú : ROC-AUC Score

## 1Ô∏è‚É£ ROC CurveÎûÄ?

ROC(Receiver Operating Characteristic) CurveÎäî  
Î∂ÑÎ•ò ÏûÑÍ≥ÑÍ∞í(threshold)ÏùÑ Î≥ÄÌôîÏãúÌÇ§Î©¥ÏÑú

- **YÏ∂ï**: True Positive Rate (TPR, Ïû¨ÌòÑÏú®, ÎØºÍ∞êÎèÑ)
- **XÏ∂ï**: False Positive Rate (FPR)

Ïùò Î≥ÄÌôîÎ•º ÎÇòÌÉÄÎÇ∏ Í∑∏ÎûòÌîÑÏù¥Îã§.

---

### üìå True Positive Rate (TPR)

$
TPR = \frac{TP}{TP + FN}
$

‚Üí Ïã§Ï†ú Positive Ï§ëÏóêÏÑú ÎßûÏ∂ò ÎπÑÏú®  
‚Üí **Recall(Ïû¨ÌòÑÏú®) = Sensitivity(ÎØºÍ∞êÎèÑ)**

---

### üìå False Positive Rate (FPR)

$
FPR = \frac{FP}{FP + TN}
$

‚Üí Ïã§Ï†ú Negative Ï§ëÏóêÏÑú ÏûòÎ™ª PositiveÎ°ú ÏòàÏ∏°Ìïú ÎπÑÏú®

---

## 2Ô∏è‚É£ AUCÎûÄ?

AUC (Area Under Curve)Îäî ROC Í≥°ÏÑ† ÏïÑÎûòÏùò Î©¥Ï†ÅÏù¥Îã§.

- **1Ïóê Í∞ÄÍπåÏö∏ÏàòÎ°ù Ï¢ãÏùÄ Î™®Îç∏**
- **0.5Ïù¥Î©¥ ÎûúÎç§ Î∂ÑÎ•òÍ∏∞**
- **0.5Î≥¥Îã§ ÏûëÏúºÎ©¥ Î∞òÎåÄÎ°ú ÏòàÏ∏°ÌïòÎäî Î™®Îç∏**

---


# üéØ ROC Ï∂ï ÌëúÍ∏∞ 2Í∞ÄÏßÄ Î∞©Ïãù

ROC Í∑∏ÎûòÌîÑÎäî ÍµêÏû¨ÎÇò Î∂ÑÏïºÏóê Îî∞Îùº Ï∂ï ÌëúÍ∏∞Í∞Ä Îã§Î•¥Í≤å ÎÇòÌÉÄÎÇúÎã§.



## üìå Î∞©Ïãù 1 (ÏùºÎ∞òÏ†ÅÏù∏ Î®∏Ïã†Îü¨Îãù ÌëúÍ∏∞)

- XÏ∂ï: **False Positive Rate (FPR)**
- YÏ∂ï: **True Positive Rate (TPR, Recall)**



## üìå Î∞©Ïãù 2 (ÌÜµÍ≥Ñ ÍµêÏû¨ ÌëúÍ∏∞)

- XÏ∂ï: **1 - Specificity**
- YÏ∂ï: **Sensitivity**



### üìå Specificity (ÌäπÏù¥ÎèÑ)

$
Specificity = \frac{TN}{TN + FP}
$

Í∑∏Î¶¨Í≥†

$
1 - Specificity = \frac{FP}{FP + TN} = FPR
$

üëâ Ï¶â, Îëê Í∑∏ÎûòÌîÑÎäî **ÏàòÌïôÏ†ÅÏúºÎ°ú ÎèôÏùºÌïú Í∑∏ÎûòÌîÑ**Ïù¥Îã§.
