### ✅ 실습 문제 11  
아래와 같은 파일 권한이 있을 때, 소유자/그룹/기타 사용자의 권한이 각각 무엇인지 설명하시오.

```text
-rw-r--r-- 1 user user 1048 memo.txt
```

**👉 예시 답안**
- 소유자(user): `rw-` → **읽기 + 쓰기 가능, 실행 불가**  
- 그룹(group): `r--` → **읽기만 가능**  
- 기타(other): `r--` → **읽기만 가능**  

---

### ✅ 실습 문제 12  
`script.sh`라는 파일에 **소유자(user)** 에게 실행 권한을 추가하시오.

**👉 예시 답안**
```bash
chmod u+x script.sh
```

---

### ✅ 실습 문제 13  
`script.sh`의 전체 권한을 `755`로 설정하시오.  
(소유자: rwx, 그룹: r-x, 기타: r-x)

**👉 예시 답안**
```bash
chmod 755 script.sh
```

---

### ✅ 실습 문제 14  
`secret.key`라는 파일을 생성하고, **소유자만 읽기·쓰기 가능** 하도록 권한을 설정하시오.

**👉 예시 답안**
```bash
touch secret.key
chmod 600 secret.key
```

---

### ✅ 실습 문제 15  
`myfolder` 디렉터리와 그 안의 모든 파일/폴더의 **소유자와 그룹을 현재 사용자로 변경**하시오.

**👉 예시 답안**

### ✅ 실습 문제 27

### ✅ 실습 문제 29  
현재 설정된 UFW 방화벽 룰을 **번호와 함께 출력**하는 명령어를 작성하시오.

**👉 예시 답안**
```bash
sudo ufw status numbered
```

---

### ✅ 실습 문제 30  
다음 작업을 순서대로 수행하는 명령어를 작성하시오.  
1. `8000/tcp` 포트를 허용한다.  
2. UFW 룰 목록을 번호와 함께 확인한다.  
3. 방금 추가한 8000 포트 룰(예: 3번)을 삭제한다.

**👉 예시 답안**
```bash
sudo ufw allow 8000/tcp
sudo ufw status numbered
sudo ufw delete 3    # 예: 3번이 8000/tcp 룰일 경우
```

---

### 🚀 추가 심화 문제 예시 답안 (아이디어)

1. SSH 서비스가 동작 중인지 확인하고, 동작하지 않으면 시작:

```bash
systemctl status ssh
sudo systemctl start ssh
```

2. SSH가 비정상 종료되었을 때, 로그로 원인 추적:

```bash
sudo journalctl -u ssh -f
sudo journalctl -u ssh --since "10 minutes ago"
```

3. 특정 IP만 SSH 접속 허용 (예: 1.2.3.4 만 허용):

```bash
sudo ufw deny 22
sudo ufw allow from 1.2.3.4 to any port 22 proto tcp
```

(실 서버 환경에서는 네트워크 구조와 운영 정책을 반드시 고려해야 함)
