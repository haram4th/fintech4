# 간단한 그래프 예제

```python
fig, ax = plt.subplots() # 그래프를 그릴 도화지 준비
ax.plot([1,2,3,4], [1,4,2,3])  # X, Y
plt.show()
```

* figure: 전체 그래프가 그려지는 도화지
* axes: 한 칸에 그려지는 그래프
* axis: 그래프의 X, Y축 눈금

# 2X2 그래프에서 왼쪽 행 결합하기
* subplot_mosaic([left, right], [left, right])

```python
fig, axes = plt.subplot_mosaic([['left', 'right_top'],
                               ['left', 'right_bottom']])
plt.show()
```

```python
fig, axes = plt.subplot_mosaic([['left', 'right'],
                               ['left', 'right']])
plt.show()
```

```python
fig, axes = plt.subplot_mosaic([['top', 'top'],
                               ['left_bottom', 'right_bottom']])
plt.show()
```

# matplotlib의 2가지 코딩 스타일

## 00-스타일: matplotlib 공식 추천 스타일
* 복잡한 그래프를 그릴 때 사용
* 사용법이 pyplot스타일보다 복잡함

```python
x = np.linspace(0, 2, 100)
x
```

```python
fig, ax = plt.subplots(figsize=(5, 2.7))
print(fig)
print(ax)
ax.plot(x, x, label="linear")
ax.plot(x, x**2, label="quadratic")
ax.plot(x, x**3, label='cubic')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('간단 그래프')
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots(3, 1, figsize=(5, 12))
print(fig)
print(ax)
ax[0].plot(x, x, label="linear")
ax[1].plot(x, x**2, label="quadratic")
ax[2].plot(x, x**3, label='cubic')
ax[0].set_xlabel('x label')
ax[0].set_ylabel('y label')
ax[0].set_title('간단 그래프')
ax[0].legend()
ax[1].set_xlabel('x label')
ax[1].set_ylabel('y label')
ax[1].set_title('quadratic 간단 그래프')
ax[1].legend()
ax[2].set_xlabel('x label')
ax[2].set_ylabel('y label')
ax[2].set_title('cubic 간단 그래프')
ax[2].legend()
plt.tight_layout()
plt.show()
```
